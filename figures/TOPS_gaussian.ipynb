```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# Define the x-axis range
x = np.linspace(-10, 10, 1000)

# Define two Gaussian distributions
y1 = np.exp(-x**2 / (2 * 4**2))  # Lower and wider
y2 = np.exp(-x**2 / (2 * 1**2)) * 2  # Higher and narrower

# Use xkcd style
with plt.xkcd(scale=1, length=200, randomness=3):
    # Plot the Gaussians
    fig, ax = plt.subplots()
    ax.plot(x, y1, color='gray', linestyle='-', linewidth=2)
    ax.plot(x, y2, color='black', linestyle='-', linewidth=2)

    # Remove the box around the plot
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

    # Add labels and ticks
    ax.set_yticks([])
    ax.set_xticks([])
```

```python
fig.savefig('gaussian_plot_xkcd.svg', bbox_inches='tight')
```
