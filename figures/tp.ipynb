```python
import numpy as np
```

```python
n_loss = np.arange(0, 5.25E-3, 2.5E-4)
```

```python
print(n_loss)
```

```python
loss_1 = np.array([0, 1.25E-5, 2.51E-4, 3.76E-4, 5.02E-4, 6.27E-4, 7.73E-4, 8.78E-4, 1E-3, 1.1E-3, 1.25E-3,
                   1.23E-3, 9.76E-4, 8.79E-4, 8.20E-4, 7.78E-4, 7.47E-4, 7.24E-4, 7.05E-4, 6.89E-4, 6.76E-4])
loss_2 = np.array([0, 1.25E-5, 2.51E-4, 3.76E-4, 5.02E-4, 6.27E-4, 7.73E-4, 8.78E-4, 1E-3, 1.1E-3, 1.25E-3,
                   1.53E-3, 2.03E-3, 2.4E-3, 2.7E-3, 3.0E-3, 3.27e-3, 3.5E-3, 3.8E-3, 4.08E-3, 4.3E-3])
```

```python
highlight_index = np.where(n_loss == 0.0025)[0][0]
```

```python
# Filter out the value where n_loss=0.0025
filtered_indices = np.arange(len(n_loss)) != highlight_index
```

```python
import matplotlib.pyplot as plt


plt.rc('text', usetex=True)
plt.rc('font', family='serif')  # Matches the default Computer Modern font in LaTeX
plt.rcParams.update({
    'font.size': 10,                # Base font size
    'axes.titlesize': 8,           # Title size
    'axes.labelsize': 8,           # Axis label size
    'xtick.labelsize': 8,           # X-tick label size
    'ytick.labelsize': 8,           # Y-tick label size
    'legend.fontsize': 8,           # Legend size
})


with plt.style.context(['science', 'ieee']):

    fig, ax = plt.subplots(figsize=(5.25,1.5))

    # Scatter plot for loss_1 and loss_2
    ax.scatter(n_loss[filtered_indices][1::2], loss_1[filtered_indices][1::2], color='blue', s=50, edgecolor='black', alpha=1, facecolor=(0, 0, 1, 0.5))
    ax.scatter(n_loss[filtered_indices][1::2], loss_2[filtered_indices][1::2], color='red', s=50, edgecolor='black', alpha=1, facecolor=(1, 0, 0, 0.5))
    ax.scatter(n_loss[filtered_indices][1:10:2], loss_2[filtered_indices][1:10:2], color='purple', s=50, edgecolor='black', alpha=1, zorder=4)


    # Line plot for loss_1 and loss_2
    ax.plot(n_loss[1:], loss_1[1:], color='blue', alpha=0.5, linestyle="-", linewidth=2)
    ax.plot(n_loss[1:], loss_2[1:], color='red', alpha=0.5, linestyle="-", linewidth=2)

    # Highlight the marker for n_loss=0.0025
    ax.scatter(n_loss[highlight_index], loss_1[highlight_index], color='yellow', marker='*', s=200, edgecolor='black', alpha=1, label='Loss 1 Highlight', zorder=5)
    ax.scatter(n_loss[highlight_index], loss_2[highlight_index], color='yellow', marker='*', s=200, edgecolor='black', alpha=1, label='Loss 2 Highlight', zorder=5)

    ax.set_yticks([0.002, 0.004, 0.006])  # Adjust the number of ticks as needed
    ax.set_xticks([0.002, 0.004, 0.006])  # Adjust the number of ticks as needed
    ax.set_xlim(-0.0005, 0.01)
    ax.set_ylim(-0.0005, 0.006)

    # Add labels and title
    #plt.xlabel('$\\Im\\{n\\}$')
    #plt.ylabel('$\\Im\\{n_\\text{eff}\\}$')
fig
```
